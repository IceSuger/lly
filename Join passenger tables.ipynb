### 先导包，并读入数据

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("D:/code/lly/CreateEdge/flight_edge/000000_0", delimiter='\t', header=None)
df2 = pd.read_csv("D:/code/lly/CreateEdge/pnr_edge/000000_0", delimiter='\t', header=None)
```

### 删掉没用的列

```python
df1 = df1.drop(2, axis=1)
df2 = df2.drop(2, axis=1)
```

### 用outer join来连接这两张表

```python
df = pd.merge(df1, df2, how='outer', on=[0, 1])
df
```

### 取两表中对应列的平均值，当有`np.nan`即空值存在时，空值会被忽略。当俩值都存在时就是求平均值。  
### 这样得到的5，6两列即为所求。

```python
df[5] = df[['3_x','3_y']].mean(1)
df[6] = df[['4_x','4_y']].mean(1)
df
```

### 删掉`3_x, 3_y, 4_x, 4_y`列即可

```python
df = df[[0,1,5,6]]
df
```
